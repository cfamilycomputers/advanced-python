### E.g7:- Write a Python Script to insert Multiple rows in a ‘emp’ table of ‘cfamilyit database?

```python
from mysql import connector
conn = connector.connect(host='localhost',
                        database='cfamilyit',
                        user='root',
                        password='Cfamily*123')
sql = '''
        insert into emp(empno,ename,sal,gneder,dno) 
        values(%s,%s,%s,%s,%s)
'''
list_of_employees = [
    (102,'swathi',15000,'f',11),
    (103,'badri',35000,'m',12),
    (104,'alia',25000,'f',12),
    (105,'pavan',25000,'m',13)
]
cursor = conn.cursor()
cursor.executemany(sql,list_of_employees)
conn.commit()
cursor.close()
conn.close()
```

### E.g8:- Write a Python Script to insert Multiple rows Dynamically into ‘emp’ table of ‘cfamilyit database?

```python
from mysql import connector
conn = connector.connect(host='localhost',
                        database='cfamilyit',
                        user='root',
                        password='Cfamily*123')
sql = '''
        insert into emp(empno,ename,sal,gneder,dno) 
        values(%s,%s,%s,%s,%s)
'''
list_of_employees = []
ch = 'y'
while ch == 'y':
    empno = int(input("Enter Empno:"))
    ename = input("Enter Ename:")
    sal = int(input("Enter Salary:"))
    gender = input("Enter gender:")
    dno = int(input("Enter dno:"))
    emp = (empno,ename,sal,gender,dno)
    list_of_employees.append(emp)
    ch = input("Do want insert one more say y for insert n for not interested:")
cursor = conn.cursor()
cursor.executemany(sql,list_of_employees)
conn.commit()
cursor.close()
conn.close()
```

### E.g10:- Select multiple rows from emp table of cfamily database?

```python
from mysql import connector
conn = connector.connect(host='localhost',
                        database='cfamilyit',
                        user='root',
                        password='Cfamily*123')

sql = "select * from emp"
cursor = conn.cursor()
cursor.execute(sql)

#fetch specified no.of records from emp table
#fetchmany(n) – To read specified no.of records from result.
rows = cursor.fetchmany(5)
for row in rows:
    print(row)
    print("*"*20)

cursor.close()
conn.close()
```

### E.g11:- Select all rows of a emp table of cfamily database?

```python
from mysql import connector
conn = connector.connect(host='localhost',
                        database='cfamilyit',
                        user='root',
                        password='Cfamily*123')

sql = "select * from emp"
cursor = conn.cursor()
cursor.execute(sql)

#fetch all records from emp table
##fetchall() – To read all records from result.
rows = cursor.fetchall()
for row in rows:
    print(row)
    print("*"*20)

cursor.close()
conn.close()
```
